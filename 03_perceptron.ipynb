## <snap style="color:red"> How weights and biases are assigned: </snap>

**1- perceptron trick:**

> Randomly assigned the values of weights and biases. That creates a line which differentiates the two regions. 

> One region is positive and the other is negative.

> Then run a loop. In every iteration, a single random data point is checked if it is in right region or wrong one

> If data point is in righ region then weights and biases remain same, otherwise weights and biases are changed.

> how many loops. 1. run 1000, 10000 times. or run convergence/ means everytime it runs check how many data points are classified. 

### <span style="color:red">How the lines is transformed:</span>

> Changing the value of C (bias), the line moves parallerly.

> Changing the value of x weigh, then line will rotate around the x-axis. 

> Changing the value of y weigh, then line will rotate around the y-axis. 

**Journey:**

    A loop runs (epochs) that checked if a point is in positive region or negative region. For example, the point is actually in positive region but according to the model the point should be in negative region. So, now the model will try transform its line so that the point can lie in positive region according to the model. It will change the value of w to transform the regions. **W new = W old - (learning rate times) * x metrics.**   

    If the positive point is in negative region then. **W new = W old + learning rate times) * x metrics**

### <span style="color:red"> **Algorithm:** </span>

for Xi 系N and wixi >= 0

    Wn  = Wo - (learning rate times)*xi 

for xi 系P and wixi<0

    wn = Wo + (learning rate times)*xi

Wn = new weight

Wo = old weight

wi = any weight

xi = any data point

系N = present in negative region

系P = present in positive regions

### <span style="color:red">**Simplified algorithm:**

Wn = Wo + (learning rate times)*(yi - y^i) * xi

y^i = output after applying function on z by the model

yi = actual output. 

How new methods works?

yi | y^i | yi-y^i | explain | region
---|-------|-----|----|---
1  |  1    |-1 | both said yes | positive in positive
0  |  1    | -1 | model no, actual yes | negative point in positive region
1  |  0    | 1 | actual no, model yes | positive point in negative region
0  |  0    | 0 | both says no | negative in negative

if the difference is zero then Wn = Wo


### <snap style="color:red"> Issue/problems with perceptron trick: </snap>

- We are not sure that the values of weights and biase is perfect of not. 
- There could be more than one lines that perfectly differentiate the regions. Can't tell which one is better.
- Can't tell how much the results are good. Can't quantify the result. 
- As we take the data points randomly, most probably we can take rightly classified data point. 

**To overcome these problems, we use loss function**

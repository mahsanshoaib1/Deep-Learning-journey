### <snap style="color:red"> Perceptrons and Neurons: </snap>

It is an algorithm/mathemetical model used for suprevised machine learning. Its 
just designed as it is look like deep learning. How it looks like:

- weights of inputs and biase to submission block.
- Submission block make dot product. X1*W1 + X2 * W2....+b
- x = the datapoint. as we always use numeral data. W is the weight of each variable. 
- z = X1*W1 + X2 * W2....+b. Then the output z to an activation function. 
- Activation function gives output
  
Thats called perceptron.

> First we train the perceptron to find the values of weights W and Bias b. 

> Weight is actually the strength of a variable. Means, how much a variable effects the output. 

> For example, to take admission cgpa is more important than iq. just like included percentage of metric
> and fsc for bs admission. 20 percent metric 40 percent fsc, remaining is the test percentage. 


### <snap style="color:red"> Geometric intuition: </snap>

> z = w1x1 + w2x2 ... + b

> y = f(x)  i.e function on x. 
 
> z >=0 then its 1

> z <=0 then its 0

> perceptron is nothing just a line. linear regression. according to the formula.

> w1 = A, w2 = B, (x1,x2=x), b=C. Then the formula is Ax+Bx+C. That is linear regression

> This line descriminate two regions in a graph. Because of two its also called **binary classifier**

> If we have three dimensional data, then this line acts as a plane.  

> 2d -> line. 3d -> plane. 4d -> hyperplane

> A drawback, it only classify linear data. For non-linear data. it doesn't work. 

## Practical

### Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading Dataset

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
X = df[["age", "fare"]]
y = df.survived
# y = y.replace("male", 1)
# y = y.replace("female", 2)
# y = y.astype("int64")
```

```python
X = X.dropna()
y = y.dropna()
X = X.iloc[ :713]
y = y.iloc[:713]
print(X.shape, y.shape)
```

### plotting Dataset 

This data is not linear but we will apply perceptron just for the sake of practice. Perceptron only works on linear data and differentiate 2 classes. For example, one region will contain survived which is 1 and other are dead which is 0. Perceptron will draw a line between them. 

```python
sns.scatterplot(x = X["age"], y = X["fare"], hue=y, palette="Set1")
```

### Applying the preceptron

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X, y) # initially it was giving an error because of missing values in data
```

```python
p.coef_
```

<snap style="color:skyblue"> The coefficents of perceptron are w1 and w2. 

> w1 = -61.68, and w2 = 57.8718

```python
p.intercept_
```

<snap style="color:skyblue"> The intercept of perceptron is the bias b. 

> b = -51

### Plotting the model

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```
